```python
class Chocolate:
  def __init__(self, price, topping):
      if price <= 0:
         raise ValueError('Цена батончика должна быть больше нуля')
      self.price = float(price)
      self.topping = topping

  def show_my_chocolate(self):
        if self.topping:
           print (f'Шоколадный батончик, добавка:{self.topping}')
        else:
             print("Обычный шоколадный батончик")

  def __str__(self):
      return f'Шоколадный батончик, {self.price} рублей'

choc1 = Chocolate(50.678, 'орехи')
choc1.show_my_chocolate()
print (choc1)
choc2 = Chocolate(30, None)
choc2.show_my_chocolate()
print(choc2)

```
